````python
#b) Create two class variables
'''
i. **ANNUAL_BONUS_RATE**: This should be applied to the 'Pay' when eligible. It should be 10%
    ANNUAL_BONUS_RATE = 0.10  # 10%
    
ii. **total_employees**: Class variable to keep track of the total number of employees created. Initial value should be zero.
'''



class Employee: 
    ANNUAL_BONUS_RATE = 0.1
    total_employees = 0
    def __init__(self, fname, lname, ID, pay, vac_days):
        self.fname = fname
        self.lname = lname
        self.ID = ID
        self.pay = pay
        self.vac_days = vac_days
        Employee.total_employees += 1
        
````
